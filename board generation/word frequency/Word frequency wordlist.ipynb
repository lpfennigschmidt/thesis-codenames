# Word frequency wordlist
Experimenting how to get wordlists with different frequencies, and whether word frequency has an effect on game performance.

Word frequency measure will be based on the SUBTLEX corpus for word frequencies as this has been found to be more predictive of common language word frequencies. Is generated from film and series subtitles.
<https://www.ugent.be/pp/experimentele-psychologie/en/research/documents/subtlexus>

The original wordlist for which we want to find frequencies will be the categories.json / default.json (all the words without their respective categories).

*Does that even make sense though? I'd imagine all of those words are quite highly frequent in everyday language, so maybe I'll have to pick out some other words in the same categories (wordnet/conceptnet) that are not as frequent*

```python
# filtering out only nouns first, using 'SUBTLEX-US frequency list with PoS and Zipf information.csv'
import pandas as pd
subtlex = pd.read_csv('SUBTLEX-US frequency list with PoS and Zipf information.csv', delimiter = ';')
subtlex
```

```python
subtlex_nouns = subtlex[subtlex['Dom_PoS_SUBTLEX'] == 'Noun']
subtlex_nouns
```

```python
# columns that are not needed
subtlex_nouns.drop(columns = ['FREQcount', 'CDcount', 'FREQlow', 'Cdlow', 'Dom_PoS_SUBTLEX', 'Lg10WF'], inplace = True)
subtlex_nouns
# lg10CD is supposed to be a good measure for ''
# Zipf-values: "Zipf values range from 1 to 7, 
# with the values 1-3 indicating low-frequency words (with frequencies of 1 per million words and lower) and 
# the values 4-7 indicating high-frequency words (with frequencies of 10 per million words and higher)"
```

```python
# load default.json, join with subtlex_nouns
import json
with open('../category norms/default.json') as file:
    default_wordlist = json.load(file)["words"]

default_df = pd.DataFrame(default_wordlist, columns = ['Word']).set_index('Word')
len(default_df)
```

```python
default_with_freq = subtlex.join(default_df, on='Word', how= "right")
default_with_freq
```

```python
default_with_freq[default_with_freq['SUBTLWF'].isna()]['Word']
```

```python
default_with_freq.dropna(inplace = True)
default_with_freq
```

```python
default_with_freq['Zipf-value'].round(0).value_counts()
```

```python
# 1-3: low freq
# 4-7: high freq
low_freq = default_with_freq[default_with_freq['Zipf-value'] < 3.5]
high_freq = default_with_freq[default_with_freq['Zipf-value']>= 4.5]
```

```python
len(low_freq)
```

```python
len(high_freq)
```

```python
MAX = min(len(high_freq), len(low_freq))
```

Because the list of high-frequency words is only half as long as the list of low-frequency words, I will shorten the list of low-frequency words to be only the same amount, taking the far end of the lower frequency.

```python
low_freq = default_with_freq.sort_values('Zipf-value').head(MAX)
low_freq
```

```python
with open('low_frequency.json', 'w') as file:
    json.dump({"words": list(low_freq['Word'])}, file)
```

```python
with open('high_frequency.json', 'w') as file:
    json.dump({"words": list(high_freq['Word'])}, file)
```

Additional gratis medium-frequency-wordlist:

```python
medium_freq = default_with_freq.loc[ abs(4 - (default_with_freq['Zipf-value'])).sort_values().index].head(MAX)
medium_freq
```

```python
with open('medium_frequency.json', 'w') as file:
    json.dump({"words": list(medium_freq['Word'])}, file)
```

# New Word Frequency List without the Category Norms
My initial problem with only taking Subtlex was that there where names and proper nouns in there as well, which do not really make sense on a codenames board. The concreteness corpus (which was build from subtlex) has a better selection of words, so I will use that one to select good words from the subtlex corpus again

```python
import pandas as pd
subtlex = pd.read_csv('SUBTLEX-US frequency list with PoS and Zipf information.csv', delimiter = ';')
subtlex
```

```python
subtlex_nouns_and_names = subtlex[subtlex['Dom_PoS_SUBTLEX'] == 'Noun']
subtlex_nouns_and_names
```

```python
concreteness_df = pd.read_csv('../word concreteness/word concreteness.csv', delimiter = ';')
concreteness_df
```

```python
subtlex_nouns = concreteness_df[['Word']].merge(subtlex_nouns_and_names, on='Word', how= "inner")
subtlex_nouns
```

```python
subtlex_nouns.sort_values('SUBTLWF')
```

```python
import json
with open('low_frequency.json', 'w') as file:
    json.dump({"words": list(subtlex_nouns.sort_values('SUBTLWF').head(250)['Word'])}, file)
```

```python
with open('high_frequency.json', 'w') as file:
    json.dump({"words": list(subtlex_nouns.sort_values('SUBTLWF').tail(250)['Word'])}, file)
```
